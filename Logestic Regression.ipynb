```python
# importig libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing Dataset
df=sns.load_dataset("iris")
df.head()
```

```python
df["species"].unique()
```

```python
df.isnull().sum()
```

```python
df=df[df["species"]!="setosa"]
```

```python
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
print(x)
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y=le.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'], 'C':[1,2,3,4,5,6,7,8,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter, scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
y_predict=classifier_regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_predict,y_test)
print(score)
```

```python
classification_report(y_predict,y_test)
```

```python
sns.kdeplot(y_predict-y_test)
sns.rugplot(y_predict-y_test)
```
