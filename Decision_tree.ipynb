```python
# importing librares
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing dataset
df=sns.load_dataset('iris')
```

```python
df
```

```python
# separting independent and dependent varaible
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
print(x)
```

```python
# cheacking wheather data value is null
df.isnull().sum()
```

```python
# Encoding last column
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y=le.fit_transform(y)
```

```python
print(y)
```

```python
# separting into train and test Dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size=0.30 , random_state=0)
```

```python
# standardizing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)

x_test=sc.transform(x_test)
```

```python
# model creation of Decision Tree
from sklearn.tree import DecisionTreeClassifier
treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(x_train,y_train)
```

```python
# Decision tree :: checking value of entropy, gini impurity, information gain
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
# final value 
y_pred=treemodel.predict(x_test)
```

```python
y_pred
```

```python
# model accuracy
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
```

```python
# accuracy score
score
```

```python
print(classification_report(y_pred,y_test))
```

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
    
}
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(x_test)
```
